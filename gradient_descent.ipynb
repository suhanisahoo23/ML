# Gradient Descent from Scratch

```
import numpy as np

X = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

m, b = 0, 0
lr = 0.01
n = len(X)

for _ in range(1000):
    y_pred = m*X + b
    dm = (-2/n)*sum(X*(y-y_pred))
    db = (-2/n)*sum(y-y_pred)
    m -= lr*dm
    b -= lr*db

print("m:", m, "b:", b)
```
