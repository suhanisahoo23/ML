```python
import sqlite3

def collect_data_from_sql(database_file, query):
    """
    Connects to a SQLite database, executes a query, and returns the results.

    Args:
        database_file (str): The path to the SQLite database file.
        query (str): The SQL SELECT statement to execute.

    Returns:
        list: A list of tuples, where each tuple represents a row of data.
              Returns an empty list if no data is found or an error occurs.
    """
    try:
        # Establish a connection to the database
        conn = sqlite3.connect(database_file)
        cursor = conn.cursor()

        # Execute the query
        cursor.execute(query)

        # Fetch all the results
        results = cursor.fetchall()

        return results

    except sqlite3.Error as e:
        print(f"Error connecting to or querying the database: {e}")
        return []

    finally:
        # Close the connection
        if conn:
            conn.close()

# Example Usage:
if __name__ == "__main__":
    # Create a dummy database and table for demonstration
    conn_setup = sqlite3.connect("example.db")
    cursor_setup = conn_setup.cursor()
    cursor_setup.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY,
            name TEXT,
            email TEXT
        )
    ''')
    cursor_setup.execute("INSERT INTO users (name, email) VALUES ('Alice', 'alice@example.com')")
    cursor_setup.execute("INSERT INTO users (name, email) VALUES ('Bob', 'bob@example.com')")
    conn_setup.commit()
    conn_setup.close()

    # Define the database file and the SQL query
    db_file = "example.db"
    sql_query = "SELECT name, email FROM users WHERE id > 0;"

    # Collect data
    data = collect_data_from_sql(db_file, sql_query)

    # Print the collected data
    if data:
        print("Collected data:")
        for row in data:
            print(row)
    else:
        print("No data collected.")
```
